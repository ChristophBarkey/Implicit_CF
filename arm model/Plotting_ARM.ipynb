```python
results_t = pd.read_csv('arm_tuning_t.csv')
```

```python
results_t['relevant_items'] = results_t.precision * results_t.num_recs
```

```python
results_t.pivot(columns='supp', index='conf', values='precision')
```

```python
fig, ax = plt.subplots(figsize=(25, 10), nrows=1, ncols=2)
names = ['precision', 'num_recs']
data = [results_t.pivot(columns='supp', index='conf', values='precision'), results_t.pivot(columns='supp', index='conf', values='num_recs')]
c = 0
for i in range(2):
    data_filtered = data[c]
    ax[i].plot(data_filtered, linestyle='-', marker='o', linewidth=2)
    ax[i].set_title(names[c], fontsize=30)
    ax[i].set_xlabel('Min Confidence', fontsize=25)
    ax[i].tick_params(axis='both', which='major', labelsize=17)
    ax[i].tick_params(axis='both', which='minor', labelsize=17)
    ax[i].autoscale()
    ax[i].grid(linestyle=':')
    c += 1
fig.legend(data_filtered.columns, loc='center right', ncol=1, title='Min Support',fancybox=True, shadow=False, frameon=True, title_fontsize=25, fontsize=20,
bbox_to_anchor=(0.94, 0.79))
#plt.savefig('curves_eals.pdf')
plt.show()
```
